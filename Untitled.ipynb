```python
text = '나는 AI 로봇!: 나는 AI 자동차, 로봇을 만드는 AI 로봇 입니다.'
skips = ['.', ',', ':', '!']
for ch in skips:
    text = text.replace(ch, '')
print(text)
```

```python
def removelse(text):
    skips = ['.', ',', ':', '!']
    for ch in skips:
        text = text.replace(ch, '')
    return text
text = '나는 AI 로봇!: 나는 AI 자동차, 로봇을 만드는 AI 로봇 입니다.'
text2 = removelse(text)
print(text2)
def max_counts(text):
    counts = {}
    for i in text.split(' '):
        if i in counts:
            counts[i] += 1
        else:
            counts[i] = 1
    return counts
text3 = max_counts(text2)
print(text3)
```

```python
def max_count(people):
    counts = {}
    for i in people:
        if i in counts:
            counts[i] += 1
        else:
            counts[i] = 1
    return counts
people = ['홍', '홍', '김', '이', '홍', '김']
counts = max_count(people)
print(counts)
max(counts.values())
```
